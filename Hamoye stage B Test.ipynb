```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
data = pd.read_csv('C:\\Users\\USER\\Documents\\python\\Homoye Internship\\energydata_complete.csv')
data.head()
```

```python
print('shape:', data.shape)
print('desribe:')
data.describe() #fuel.describe(include='all') to include all
```

```python
data.info()
```

```python
data = data.drop(columns =['lights', 'date'])
data.head()
```

```python
from sklearn.preprocessing import MinMaxScaler
scaler = MinMaxScaler()
normalised_data = pd.DataFrame(scaler.fit_transform(data), columns = data.columns)
normalised_data.head()
```

```python
features = normalised_data.loc[:,'T1': 'rv2']
features.head()
```

```python
target = normalised_data['Appliances']
target.head()
```

```python
from sklearn.model_selection import train_test_split 
```

```python
from sklearn.linear_model import LinearRegression
linear_model = LinearRegression()
linear_model
```

```python
print('Intercept:', linear_model.intercept_)
```

```python
coef = pd.DataFrame({'features': features.columns, 'weights': linear_model.coef_}).sort_values('weights').reset_index()
coef = coef.drop('index', axis =1)
coef
```

# 

```python
temp_linear = linear_model.fit(normalised_data[['T2']], normalised_data[['T6']])
```

```python
temp_predict = linear_model.predict(normalised_data[['T2']])
```

```python
# R-Squared (r2)
from sklearn.metrics import r2_score
r2_value = r2_score(normalised_data[['T6']], temp_predict)
round(r2_value, 2)
```

```python
# Mean Absolue Error (mae)
from sklearn.metrics import mean_absolute_error
mae = mean_absolute_error(normalised_data[['T6']], temp_predict)
round(mae, 2)
```

```python
# Residual Sum of Squares (RSS)
t6 = normalised_data[['T6']]
rss = np.sum(np.square(t6 -temp_predict))
round(rss, 2)
```

```python
# Root Mean Square Error (RMSE)
from sklearn.metrics import mean_squared_error
rmse = np.sqrt(mean_squared_error(normalised_data[['T6']], temp_predict))
round(rmse, 3)
```

```python
coef = linear_model.coef_
coef.round(2)
```

```python
x_train, x_test, y_train, y_test = train_test_split(features, target, test_size = 0.3, random_state = 42)
```

```python
linear_reg = linear_model.fit(x_train, y_train)
linear_reg
```

```python
linear_predict = linear_reg.predict(x_test)
linear_predict
```

```python
from sklearn.linear_model import Ridge
ridge_reg = Ridge(alpha =0.4)
ridge_reg.fit(x_train, y_train)
```

```python
from sklearn.linear_model import Lasso
lasso_reg = Lasso(alpha = 0.001)
lasso_reg.fit(x_train, y_train)
```

```python
# Mean Absolue Error (mae) for linear
from sklearn.metrics import mean_absolute_error
mae = mean_absolute_error(linear_predict, y_test)
round(mae, 3)
```

```python
ridge_predict = ridge_reg.predict(x_test)
# Mean Absolue Error (mae) for ridge
from sklearn.metrics import mean_absolute_error
mae = mean_absolute_error(ridge_predict, y_test)
round(mae, 3)
```

```python
print('lasso coefficients: \n', lasso_reg.coef_)
```

```python
# Root Mean Square Error (RMSE) for lasso
rmse = np.sqrt(mean_squared_error(y_test, lasso_reg.predict(x_test)))
round(rmse, 3)
```
